```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    """
    Returns velocity field generated by a vortex 
    
    strength: float, strength of vortex
    xv: float, x-coordinate of the vortex
    yv: float, y-coordinate of the vortex
    X: 2D numpy array of the vortex, x-coordinates
    Y: 2D numpy array of the vortex, y-coordinates
    
    u: 2D numpy array of floats, x-component of velocity vector field 
    v: 2D numpy array of floats, y-component of velocity vector field 
    """
    u = +strength / (2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = -strength / (2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u,v
```
